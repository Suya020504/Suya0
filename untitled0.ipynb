```python
!pip install requests beautifulsoup4 pandas openpyxl

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from urllib.parse import quote

def crawl_daisomall(keyword, max_pages=5):
    base_url = "https://www.daisomall.co.kr/ds"  # 실제 URL 구조 확인해서 수정 필요
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    results = []

    for page in range(1, max_pages + 1):
        params = {
            "keyword": keyword,
            "page": page
        }
        res = requests.get(base_url, params=params, headers=headers)
        res.raise_for_status()

        soup = BeautifulSoup(res.text, "html.parser")

        # ↓↓↓ 여기가 사이트 구조에 맞춰 수정해야 하는 부분 ↓↓↓
        product_list = soup.select(".product-item")  # 예시 CSS 선택자

        # 더 이상 상품이 없으면 중단
        if not product_list:
            break

        for item in product_list:
            name_el = item.select_one(".product-name")
            price_el = item.select_one(".product-price")
            link_el = item.select_one("a")

            name = name_el.get_text(strip=True) if name_el else None
            price = price_el.get_text(strip=True) if price_el else None
            url = "https://www.daisomall.co.kr" + link_el["href"] if link_el and link_el.get("href") else None

            results.append({
                "keyword": keyword,
                "product_name": name,
                "price": price,
                "url": url
            })

        # 서버 부하 줄이기 위해 딜레이
        time.sleep(1)

    df = pd.DataFrame(results)
    return df

# 실제 실행 예시
keyword = "살균제"  # 네가 찾고 싶은 약품 키워드
df = crawl_daisomall(keyword, max_pages=10)

# 엑셀로 저장
output_filename = f"daisomall_{keyword}_search.xlsx"
df.to_excel(output_filename, index=False)
output_filename
```
